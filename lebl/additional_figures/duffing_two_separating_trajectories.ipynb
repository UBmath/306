```python
from resources306 import *
from scipy.integrate import solve_ivp
from matplotlib.ticker import AutoMinorLocator
```

```python
def Duff(t,X):
    x,y = X
    return [y,-0.05*y-x**3+8*np.cos(t)]

T = 15

t = np.linspace(0,T,10000)

Y1 = solve_ivp(Duff,[0,T],[2,3],t_eval=t)
Y2 = solve_ivp(Duff,[0,T],[2,2.9],t_eval=t)

fig = plt.figure(figsize=(12,4))
plt.subplots_adjust(wspace=0.3)

plt.subplot(121)
ax = fig.gca()
ax.plot(Y1.y[0], Y1.y[1],'b')
ax.plot(Y2.y[0], Y2.y[1],'r')
#ax.set_title("Duffing Equation: Phase Plane")
ax.set_xlabel('x',fontsize=15)
ax.set_ylabel("x'",fontsize=15)
ax.set_yticks([-5+n*2.5 for n in range(5)])
ax.set_xticks([-2,0,2])
ax.xaxis.set_minor_locator(AutoMinorLocator(10))
ax.yaxis.set_minor_locator(AutoMinorLocator(10))
ax.tick_params(which='both',direction='in',right='true',top='true',labeltop='true',\
               labelright='true')
ax.grid(ls='--')

plt.subplot(122)
ax2 = fig.gca()
plt.plot(Y1.t,Y1.y[0],'b')
plt.plot(Y2.t,Y2.y[0],'r')
#ax2.set_title("Duffing equation: $t$-$x$-plane")
ax2.grid(ls='--')
ax2.xaxis.set_minor_locator(AutoMinorLocator(10))
ax2.yaxis.set_minor_locator(AutoMinorLocator(10))
ax2.set_xticks([n*2.5 for n in range(7)])
ax2.tick_params(which='both',direction='in',right='true',top='true',labeltop='true',\
               labelright='true')
ax2.set_xlabel('$t$',fontsize=15)
ax2.set_ylabel('$x$',fontsize=15)

plt.savefig('duffing_two_separating_trajectories.svg')
plt.show()
```
